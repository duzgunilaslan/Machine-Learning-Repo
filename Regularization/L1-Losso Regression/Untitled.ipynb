```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.datasets import load_breast_cancer
%matplotlib inline
```

```python
cancer = load_breast_cancer()
print(cancer.keys())
```

```python
cancer_data = pd.DataFrame(cancer.data,columns=cancer.feature_names)
```

```python
cancer_data
```

```python
X = cancer_data
```

```python
y=pd.DataFrame(cancer.target)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y ,shuffle=True, test_size=0.2,random_state=42)
```

```python
X_train
```

```python
lasso = Lasso()
lasso.fit(X_train,y_train)
num = coeff_used00001 = np.sum(lasso.coef_!=0)
print("Train Score:", lasso.score(X_train,y_train))
print("Test Score:", lasso.score(X_test,y_test))
print("number of features used",num)
```

```python
lasso = Lasso(alpha=0.01,max_iter=10e5)
lasso.fit(X_train,y_train)
num = coeff_used00001 = np.sum(lasso.coef_!=0)
print("Train Score:", lasso.score(X_train,y_train))
print("Test Score:", lasso.score(X_test,y_test))
print("number of features used",num)
```

```python
lasso = Lasso(alpha=0.0001,max_iter=10e5)
lasso.fit(X_train,y_train)
num = coeff_used00001 = np.sum(lasso.coef_!=0)
print("Train Score:", lasso.score(X_train,y_train))
print("Test Score:", lasso.score(X_test,y_test))
print("number of features used",num)
```

```python
linear = LinearRegression()
linear.fit(X_train,y_train)
num = coeff_used00001 = np.sum(linear.coef_!=0)
print("Train Score:", linear.score(X_train,y_train))
print("Test Score:", linear.score(X_test,y_test))
print("number of features used",num)
```
