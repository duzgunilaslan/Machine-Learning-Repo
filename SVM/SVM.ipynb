```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv("wine.csv")
```

```python
data = pd.DataFrame(data)
```

```python
data
```

```python
# to separate the data set as a X and y
X = data.drop("Class", axis=1)
```

```python
X
```

```python
y = data.Class
```

```python
# split the dataset 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle= True, stratify= y)
```

```python
model = SVC()
```

```python
#determine the parameter values
param_grid = {"C":(1,10,100,1000),"gamma":(0.1,0.01,0.001,0.0001)}
```

```python
# create the Grid Search model
grid_search = GridSearchCV(model, param_grid, cv=10, verbose=10,return_train_score=False)
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Check the best parameter
grid_search.best_params_
```

```python
# to place the best parameters and fit
model = SVC(C=1, gamma=0.1).fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
predict = model.predict(X_test)
```

```python
#Control the confusion matrix and classification Report
print("Confusion Matrix: ")
print(confusion_matrix(y_test, predict))
print()
print("Classifiction Report :")
print(classification_report(y_test, predict))
```
