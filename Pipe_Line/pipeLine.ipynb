```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data
```

```python
data = pd.get_dummies(data,prefix ="_")
```

```python
data
```

```python
test = data.iloc[14:19,:]
```

```python
test
```

```python
dataset = data.drop(labels=range(14, 19),axis=0)
```

```python
dataset
```

```python
features = ["No","Age","__C","__Q","__S","__Famale","__Male"]  
```

```python
X , y = dataset[features], dataset["Sallary"]
```

```python
X
```

```python
y
```

```python
test = test[features]
```

```python
test
```

```python
imputer = SimpleImputer()
model = LinearRegression()
```

```python
pipeline = make_pipeline(imputer,model)
```

```python
pipeline.fit(X,y)
```

```python
pipeline.predict(test)
```
